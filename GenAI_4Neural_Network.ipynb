```python
#Step 1:Import necessary libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

#Step 2:Generate some random data for demonstration
np.random.seed(0)
X=np.random.rand(100,2)
y= (X[:,0]+X[:,1]> 1).astype(int)
y
```

```python
#Step 3: Create a Sequential model
model=Sequential()

#Step 4: Add layers to the model
model.add(Dense(3,input_dim=2,activation='relu')) #Hidden layers with 3 neurons
model.add(Dense(1,activation='sigmoid')) #Output layer with 1 neuron(Binary Classification)
```

```python
X
```

```python
y
```

```python
#Step 5 :Compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#Step 6: Train the model
model.fit(X,y,epochs=12,batch_size=25)

#Step 7:Evaluate the model
loss,accuracy=model.evaluate(X,y)
print(f'Loss:{loss:.4f}')
print(f'Accuracy:{accuracy*100:.2f}%')
```

```python
#Implement using Dataset for multiclassification
#Step 1: Import necessary libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
#Step 2: Load and preprocess dataset
iris=load_iris()
X=iris.data
y=iris.target
y_binary=np.where(y==0,0,1) #Converting to binary classification

#One Hot encoding the binary labels
y_binary=to_categorical(y_binary)

#Split data into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y_binary,test_size=0.2,random_state=42)
```

```python
feature_names=iris.feature_names
print("Feature Names: ",feature_names)
import pandas as pd
Features=pd.DataFrame(X)
Features
```

```python
#Step 3:Create a sequential model
model=Sequential()

#Step 4: Add layers to the model
model.add(Dense(8,input_dim=4,activation='relu')) #Hidden layers with 8 neurons
model.add(Dense(2,activation='softmax')) #Output layer with 3 neuron(Multi-Classification)

#Step 5: Compiling the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

#Step 6: Train the model
model.fit(X_train,y_train,epochs=100,batch_size=20)
```

```python
#Step 7: Evaluate the model
loss,accuracy=model.evaluate(X_test,y_test)
print(f'Loss:{loss:.4f}')
print(f'Accuracy:{accuracy*100:.2f}%')
```

```python
#Print the feauture names
feature_names=iris.feature_names
print("Feature Names: ",feature_names)
```

```python
#Get the learned weights from first layer
weights=model.layers[0].get_weights()[0]

#print feature importance based on learned weights
feature_importance=np.mean(np.abs(weights),axis=0)
print("Feature Importance: ",feature_importance)
print("Accuracy: ",accuracy)
```

```python
#Identify the index of most important feature
most_important_feature_index=np.argmax(feature_importance)
print("Index of most important feature: ",most_important_feature_index)
```
